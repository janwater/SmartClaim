```python
import os
def _create_path(path):
    if not os.path.exists(path):
        os.mkdir(path)       

def _create_filename(path, fname, prefix = '', affix = '.csv'):
    # create new file name
    output_name = prefix + fname + affix        
    foutput_name = os.path.join(path, '', output_name)
             
    return foutput_name
```
